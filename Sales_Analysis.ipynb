```python
import pandas as pd
```

```python
walmart_clean=pd.read_csv('Walmart.csv')
walmart_clean.head()
```

```python
walmart_clean.isnull().sum()
```

```python
walmart_clean.duplicated().sum()
```

```python
#Removed duplicate values
walmart_clean.drop_duplicates(inplace=True) 
walmart_clean.head()
```

```python
walmart_clean.describe()
```

```python
#dropping rows with null values
walmart_clean.dropna(inplace=True)
```

```python
walmart_clean.describe()
```

```python
walmart_clean.head()
```

```python
walmart_clean['unit_price']= walmart_clean['unit_price'].str.replace('$','').astype(float)
walmart_clean.head()
```

```python
walmart_clean.dtypes
```

```python
walmart_clean['total']=walmart_clean['unit_price'] * walmart_clean['quantity']
```

```python
walmart_clean.head()
```

```python
walmart_clean.to_csv('walmart_clean1.csv',index=False)
```
